```python
#Loading the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the Iris DataSet

df_iris = sns.load_dataset('iris')
df_iris.head()
```

```python
#Counting data for each species

df_iris['species'].value_counts()
```

```python
#Visualizing the same using a pie plot

df_iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
```

```python
#Checking the structure of the dataframe
df_iris.info()
```

```python
#checking if there are any null values in the dataset
df_iris.isnull().sum()
```

```python
#Visualization data using .plot from the pandas library

df_iris.plot(kind = 'scatter',x = 'sepal_length',y = 'sepal_width')
plt.title('Scatter plot for SepalLenght Vs. SepalWidth')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```

```python
df_iris.plot(kind = 'scatter',x = 'petal_length',y = 'petal_width')
plt.title('Scatter plot for PetalLenght Vs. PetalWidth')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid()
plt.show()
```

```python
#plotting the same visualization that is seen above using seaborn library
#we will use a jointplot to show the scatter along with the histogram.

plt.figure(figsize = (10,7),dpi = 100)
sns.jointplot(x = 'sepal_length',y = 'sepal_width',data = df_iris,size = 5)
plt.title('Joint plot for SepalLenght Vs. SepalWidth')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10,7), dpi = 100)
sns.jointplot(x = 'petal_length',y = 'petal_width',data = df_iris,size = 5)
plt.title('Joint plot for PetalLenght Vs. PetalWidth')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid()
plt.show()
```

```python
#The above scatter plots do not distinguish between the species
#Using a scatterplot from the seaborn library with hue for species we can distinguish the species for the same

plt.figure(figsize = (7,5), dpi = 100)
sns.scatterplot(x = 'sepal_length',y = 'sepal_width',hue = 'species',data = df_iris)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (7,5), dpi = 100)
sns.scatterplot(x = 'petal_length',y = 'petal_width',hue = 'species',data = df_iris)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid()
plt.show()
```

```python
#To visualize bivariate relation between each pair in the data set use pairplot function from the seaborn library.

sns.pairplot(df_iris,hue = 'species')
```

```python
#To visualize the correlation between 2 pairs we can use the heatmap function from the seaborn library
#first we have to generate the correlation matrix and then pass that to the heatmap function.

x = df_iris.corr()
x
```

```python
sns.heatmap(x)
```
